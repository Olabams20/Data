Task 1: Class Test Scores
Scores: 5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20

Mean

Total = 5 + 8 + 10 + 12 + 12 + 12 + 13 + 14 + 14 + 15 + 16 + 18 + 18 + 20 + 20 = 227
Mean = 227 / 15 = 15.13

Median
Middle value in an ordered list of 15 scores:

Median = 8th score = 14

Mode
Most frequent score(s):

12 appears 3 times  Mode = 12

If the lowest score (5) was wrongly recorded and should have been 10:
New total = 227 - 5 + 10 = 232 New mean = 232 / 15 = 15.47 (approx) Median remains 14 middle value unchanged Mode remains 12 still most frequent

Explanation:

Mean increases because the total sum increases.

Median and mode are unaffected since the change doesn’t alter the middle value or the frequency of the most common score.

Task 2: Daily Commute Times
Times (minutes): 15, 20, 20, 25, 25, 30, 60, 65, 120

Mean
Total = 15 + 20 + 20 + 25 + 25 + 30 + 60 + 65 + 120 = 380

Mean = 380 / 9 = 42.22

Median
Middle value in ordered list:

Median = 5th value = 25

Mode
Most frequent values:

20 and 25 both appear twice Mode = 20 & 25

Median is the best representation.


The mean is skewed by the extreme value (120 minutes).

The median (25 minutes) better reflects the central tendency of most students.

Mode is useful but less representative here due to dual peaks.

Task 3: Shoe Sizes
Sizes: 38, 39, 38, 40, 41, 42, 39, 39, 40, 38, 37, 39, 40, 41, 42

Mode
39 appears 4 times Mode = 39

If a shoe company wants to donate free shoes:
They should produce size 39.

It’s the most common size, maximizing the number of students who can wear the shoes.

If they want one size to fit the average student: They should use the mode (size 39)
