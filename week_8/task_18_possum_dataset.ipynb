```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data =pd.read_csv('possum.csv')
data.head()
```

```python
numeric = data.select_dtypes(include=['float64', 'int64'])
print("Numeric columns:\n", numeric.columns.tolist())
```

```python
# Mean, median, mode
means = numeric.mean()
medians = numeric.median()
modes = numeric.mode().iloc[0]  # first mode row

print("Mean:\n", means)
print("\nMedian:\n", medians)
print("\nMode:\n", modes)






```

```python
# Range
range_vals = numeric.max() - numeric.min()

# Variance and Standard Deviation
variance = numeric.var()
std_dev = numeric.std()

# Interquartile Range 
Q1 = numeric.quantile(0.25)
Q3 = numeric.quantile(0.75)
IQR = Q3 - Q1

print("Range:\n", range_vals)
print("\nVariance:\n", variance)
print("\nStandard Deviation:\n", std_dev)
print("\nIQR:\n", IQR)


```

```python
outliers = {}
for col in numeric.columns:
    Q1 = numeric[col].quantile(0.25)
    Q3 = numeric[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outlier_rows = numeric[(numeric[col] < lower) | (numeric[col] > upper)]
   
    if outlier_rows.empty:
        outliers[col] = "No outliers detected"
    else:
        outliers[col] = outlier_rows[col].tolist()

print("\nOutliers detected:\n")
for col, vals in outliers.items():
    print(f"{col} : {vals}")
```
